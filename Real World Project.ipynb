# Data Science Project by Using car Data Set

```python
# In this project I used a data set of car with some attributes to perform the following tasks
# import packages I need and read the csv file data
import pandas as pd
cars=pd.read_csv('car.csv')
cars.head()

```

```python
# check rows and columns in our data
cars.shape
```

```python
# Data Cleaning: Find all the null values
cars.isnull().sum()
```

```python
#Fill the null values of any column with the mean of that column
cars['Cylinders'].fillna(cars['Cylinders'].mean(),inplace=True)
```

```python
# check what are the different types of makes in our dataset and what is the cound(entries/rows) of each make?
cars['Make'].value_counts()
```

```python
# Show all the records based on origin in Asia(Filtering)
cars[cars['Origin'].isin(['Asia'])]
```

```python
#removing unwanted records: remove all the cars with engine grater than 3.5
cars[~(cars['EngineSize']>3.5)]
```

```python
# Increase the vaues of MPG_Highway by 5
cars['MPG_Highway']=cars['MPG_Highway'].apply(lambda x: x+5)
cars
```

```python
# Find the cars of Make Toyota
x=cars[cars.Make.str.contains('Toyota')]
x
```

```python
# Find Sedan cars of Toyota
x[x.Type.str.contains('Sedan')]
```

```python
cars.head() # check the dataset again
```

```python
# Find the SUV's of Europe with price garter than $3000?
cars[(cars.Type=='SUV')&(cars.Origin=='Europe')&(cars.MSRP>'$30000')]
```

```python
# Find the cars MPG Highway grater than 28 or engine size grater than 3.0
cars[(cars.MPG_Highway>28)|(cars.EngineSize>3.0)]
```

```python
# plot the different types of bodies in the dataset with pie chart 
cars.groupby('Type').Type.count().plot(kind='pie')
```

```python
# to draw the bar graph by using seaborn library
import seaborn as sns # import seaborn library
sns.countplot(x=cars.Type)
```

```python
cars.head() # check the dataset again to recall the entries
```

```python
# Find the price of most expensive car's of Europe ? 
cars[(cars.Origin=='Europe')].MSRP.max()
```

```python
# What is the Make of Most expensive car's of Europe?
cars[(cars.Origin=='Europe')].Make.max()
```

```python
# Find the total number of SUV's in our dataset?
cars.Type.str.contains('SUV').sum()
```

```python
# Rearrange the dataset based on price low to high
cars.sort_values(by='MSRP')
```

```python
# sort the dataset in descending order of prices
cars.sort_values(by='MSRP',ascending='False')
```
